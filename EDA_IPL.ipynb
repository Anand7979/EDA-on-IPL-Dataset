```python
highest_profit_year=ipl_data.groupby('Year')['Total in Crores'].max()
highest_profit_teams=ipl_data.groupby(['Year','Teams'])['Total in Crores'].max()


profit_year_df=pd.DataFrame(columns=['Year','Teams','Total in Crores'])
for i in highest_profit_year.iteritems():
    for j in highest_profit_teams.iteritems():
        if i[0]==j[0][0] and i[1]==j[1]:
            profit={'Year':j[0][0],
           'Teams':j[0][1],
           'Total in Crores':j[1]}
            profit_year_df=profit_year_df.append(profit,ignore_index=True)          
profit_year_df            
```

```python
highest_profit_year
```

```python
plt.title('Teams Profit')
sns.barplot('Total in Crores','Teams', hue='Year', data=ipl_data, palette='Paired')
```

```python
plt.title('Teams Maximum Profit')
sns.barplot('Total in Crores','Teams', hue='Year', data=profit_year_df, palette='Paired')
```

```python
highest_profit_broadcast=ipl_data.groupby('Year')['Broadcasting(STAR SPORTS)'].max()
highest_profit_teams_broadcast=ipl_data.groupby(['Year','Teams'])['Broadcasting(STAR SPORTS)'].max()


profit_year_df=pd.DataFrame(columns=['Year','Teams','Broadcasting(STAR SPORTS)'])
for i in highest_profit_broadcast.iteritems():
    for j in highest_profit_teams_broadcast.iteritems():
        if i[0]==j[0][0] and i[1]==j[1]:
            profit={'Year':j[0][0],
           'Teams':j[0][1],
           'Broadcasting(STAR SPORTS)':j[1]}
            profit_year_df=profit_year_df.append(profit,ignore_index=True)          
profit_year_df            
```

```python
plt.title('Teams Broadcast Right Profit')
sns.barplot('Broadcasting(STAR SPORTS)','Teams', hue='Year', data=ipl_data, palette='Paired')
```

```python
plt.title('Teams Maximum Broadcast Right Profit')
sns.barplot('Broadcasting(STAR SPORTS)','Teams', hue='Year', data=profit_year_df, palette='Paired')
```

```python
highest_profit_ticket=ipl_data.groupby('Teams')['Ticket Money'].max()
highest_profit_ticket
```

```python
ticket_df=pd.DataFrame(columns=['Teams','Ticket_Profit'])
for i in highest_profit_ticket.iteritems(): 
    profit={'Teams':i[0],
           'Ticket_Profit':i[1]}
    ticket_df=ticket_df.append(profit,ignore_index=True)
ticket_df   
```
